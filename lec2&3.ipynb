```python
#fibonacci series
n=int(input("Enter limit of fibonacci series"))
f0,f1=0,1
for i in range(0,n,1):
    if i<=1:
        print(i)
    else:
        fn=f0+f1
        print(fn)
        f0,f1=f1,fn    
```

```python
#to find whether the input is a part of fibonacci series
n=int(input("Enter limit of fibonacci series"))
find=int(input("Enter number to be found: "))
f0,f1=0,1
ifFound=False
for i in range(0,n,1):
    if i<=1:
        print(i)
    else:
        fn=f0+f1
        if fn==find:
            ifFound=True
            print(f" {fn} Found at position: {i+1} ")
            break
        print(fn)
        f0,f1=f1,fn  
if ifFound==False:
    print("Not found")        
        

```

```python
#list armstrong numbers from 100 to 999

for i in range(100,1000,1):
    sum=0
    dummy=i
    while(i>0):       #eventually after some point 0 will come and since 0 is not greater tahn 0 then the control will come out of the loop
        digit=(i%10)
        i=i//10
        sum += (digit**3)
    if dummy==sum:
        print(f"Armstrong number found: {sum}")
```
